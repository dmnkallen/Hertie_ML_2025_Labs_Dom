```python
#7
import numpy as np
#OLS function
def OLS(X, y):
    """
    Compute OLS solution using closed-form formula.
    """
    X = np.hstack((np.ones((X.shape[0], 1)), X))  
    theta = np.linalg.inv(X.T @ X) @ X.T @ y  
    return theta


```

```python
# Extract features as a 2D array
X3_vec = X_df['X3 distance to the nearest MRT station'].values.reshape(-1, 1)
X4_vec = X_df['X4 number of convenience stores'].values.reshape(-1, 1)
```

```python
# Combine features into a single matrix 
X = np.hstack((X3_vec, X4_vec))
```

```python
y = y_df['Y house price of unit area'].values
```

```python
# Compute OLS coefficients
theta = OLS(X, y)
```

```python
# Print results
print("Estimated coefficients:", theta)
```

```python
#8 statsmodel 


import statsmodels.api as sm

# Add an intercept 
X_sm = sm.add_constant(X)

# Fit the OLS model
model_sm = sm.OLS(y, X_sm).fit()

# Get the coefficients
theta_sm = model_sm.params
print("Statsmodel OLS Coefficients:", theta_sm)
```

```python
#8 scikitlearn 

from sklearn.linear_model import LinearRegression

# Initialize and fit the model
model_sklearn = LinearRegression(fit_intercept=True)
model_sklearn.fit(X, y)

# Get coefficients (Intercept + feature coefficients)
theta_sklearn = np.hstack((model_sklearn.intercept_, model_sklearn.coef_))
print("Scikit-Learn OLS Coefficients:", theta_sklearn)
```

```python
#9
import matplotlib.pyplot as plt
```

```python
# Function to generate predictions using OLS coefficients
def predict_OLS(X, theta):
    """Compute predictions using the estimated OLS coefficients."""
    X = np.hstack((np.ones((X.shape[0], 1)), X))  #Intercept
    return X @ theta  # Matrix multiplication for predictions
```

```python
# Generate predictions
y_pred = predict_OLS(X, theta)  # Using estimated coefficients from OLS function
```

```python
# Print first few predictions for reference
print("First 20 Predictions:", y_pred[:20])
print("First 20 Actual Values:", y[:20])
```

```python

plt.scatter(y, y_pred, alpha=0.7, edgecolors='k', label="Predictions")  # Scatter plot of true vs predicted
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--', label="Perfect Fit")  # Reference line
plt.xlabel("True House Prices (y)")
plt.ylabel("Predicted House Prices ()")
plt.title("True vs. Predicted House Prices (OLS)")
plt.legend()
plt.show()
```

```python
#10


def MSE(y, y_hat):
    """
    Compute the MSE between true labels and predictions.

    Parameters:
    y (numpy.ndarray): True house prices.
    y_hat (numpy.ndarray): Predictions from the model.

    Returns:
    float: Mean Squared Error.
    """
    mse = np.mean((y - y_hat) ** 2)  
    return mse
```

```python
#13

from sklearn.metrics import mean_squared_error
```

```python
# Standardizing the input feature
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
train_mse = []  
test_mse = []   
```

```python
for degree in range(11):
    # Polynomial transformation of degree `degree`
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X_scaled)

    # Train-test split (80-20)
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

    # Fit the polynomial regression model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Predict on both training and test sets
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Calculate MSE for training and testing sets
    train_mse.append(mean_squared_error(y_train, y_train_pred))
    test_mse.append(mean_squared_error(y_test, y_test_pred))
```

```python
# Plotting the train and test MSE
plt.plot(range(11), train_mse, label='Train MSE', color='blue')
plt.plot(range(11), test_mse, label='Test MSE', color='red')
plt.xlabel('Degree of Polynomial')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Train vs Test MSE for Polynomial Regression (Degree 0 to 10)')
plt.legend()
plt.show()
```

The chart above shows that at all degrees the the model fits the training data perfectly but at above 9 degrees the MSE for the test data goes way up at showing that it is overfitting to the training data.

```python
#14

# Standardizing the input feature
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

train_mse = []  # Store train MSE values
test_mse = []   # Store test MSE values

# Loop through polynomial degrees from 0 to 10
for degree in range(11):
    # Polynomial transformation of degree `degree`
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X_scaled)

    # Train-test split with a different random seed
    X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=33)

    # Fit the polynomial regression model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Predict on both training and test sets
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Calculate MSE for training and testing sets
    train_mse.append(mean_squared_error(y_train, y_train_pred))
    test_mse.append(mean_squared_error(y_test, y_test_pred))

# Find the degree that minimizes the test MSE
min_test_mse = min(test_mse)
best_degree = test_mse.index(min_test_mse)

# Print the degree and the minimum test MSE
print(f"Degree that minimizes test MSE: {best_degree}")
print(f"Minimum test MSE: {min_test_mse}")

# Plotting the train and test MSE
plt.plot(range(11), train_mse, label='Train MSE', color='blue')
plt.plot(range(11), test_mse, label='Test MSE', color='red')
plt.xlabel('Degree of Polynomial')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Train vs Test MSE for Polynomial Regression (Degree 0 to 10)')
plt.legend()
plt.show()

#used chatgpt to help create the loop for answering this question
```
